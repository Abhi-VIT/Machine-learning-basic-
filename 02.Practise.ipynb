```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
data = {
    'area':[2600, 3000, 3200, 3600, 4000],
    'price' : [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df.area, df.price)
plt.scatter(df.area, df.price)
plt.plot([df.area.iloc[0], df.area.iloc[-1]], [df.price.iloc[0], df.price.iloc[-1]], color='blue')
plt.plot([df.area.iloc[1], df.area.iloc[-2]], [df.price.iloc[0], df.price.iloc[-2]], color='red')
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
# sklearn - r2
predict = reg.predict([[3000]])
predict
```

```python
accuracy = reg.score(df[['area']], df.price)
accuracy
```

```python
lst = [91,95,54,69,80,85,88,73,71,70,66,90,86,84,73]
def bubble_sort(lst):
    n = len(lst)
    for i in range(n-1):
        for j in range(n-i-1):
            if lst[j]>lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    print(lst)

bubble_sort(lst)
```
